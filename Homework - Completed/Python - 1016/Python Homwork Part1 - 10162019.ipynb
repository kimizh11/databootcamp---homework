```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\kimi666\Desktop\Class Matericals - Piro - Git Lab\UT-TOR-DATA-PT-09-2019-U-C\Unit 3 - Python\Homework\Instructions\PyBank\Resources\budget_data.csv")
df
```

```python
Total_Month = df.shape[0]
Total_Month
```

```python
Total = df['Profit/Losses'].sum()
Total
```

```python
df['change_by_month'] = df['Profit/Losses'].diff()
df.head(30)
```

```python
Average_change = round(df['change_by_month'].mean(),2)
Average_change
```

```python
Greatest_Increase_month = df.loc[df['change_by_month']==df['change_by_month'].max()]['Date'].values[0]
Greatest_Increase_changes= df.loc[df['change_by_month']==df['change_by_month'].max()]['change_by_month'].values[0]
Greatest_Increase_month
Greatest_Increase_changes
```

```python
Greatest_Decrease_month = df.loc[df['change_by_month']==df['change_by_month'].min()]['Date'].values[0]
Greatest_Decrease_changes = df.loc[df['change_by_month']==df['change_by_month'].min()]['change_by_month'].values[0]
Greatest_Decrease_month
Greatest_Decrease_changes
```
