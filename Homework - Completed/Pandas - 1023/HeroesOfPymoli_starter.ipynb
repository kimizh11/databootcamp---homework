```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#total number of players wouldn't count duplicate names
#same player could buy more than games so we use "unique()"
#make it to a dataframe
Total_Players = len(purchase_data["SN"].unique())
playercount_total = pd.DataFrame({"Total Players": [Total_Players]})
playercount_total
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#use len() to count how many in the column
#formate the numbers with dollar sign and two decimals

Unique_Items = len(purchase_data["Item ID"].unique())
Average_Price = purchase_data["Price"].mean()
Total_Purchase = len(purchase_data["Purchase ID"])
Total_Revenue = purchase_data["Price"].sum()
Purchase_Analysis = pd.DataFrame({"Unique Items": [Unique_Items],
                                 "Average Price": [Average_Price],
                                "Total Number of Purchases": [Total_Purchase],
                                 "Total Revenue": [Total_Revenue]})
Purchase_Analysis["Average Price"]=Purchase_Analysis["Average Price"].map("${:.2f}".format)
Purchase_Analysis["Total Revenue"]=Purchase_Analysis["Total Revenue"].map("${:.2f}".format)
Purchase_Analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#use value_counts() to get the numbers of each gender
#use str() to convert the numbers to string and join it with "%"
#put all in the dataframe

Male_Players = purchase_data["Gender"].value_counts()["Male"]
Female_Players = purchase_data["Gender"].value_counts()["Female"]
NonDisclosed_Players = purchase_data["Gender"].value_counts()["Other / Non-Disclosed"]

Male_Percentage = str(round(Male_Players/Total_Purchase,4)*100) + "%"
Female_Percentage = str(round(Female_Players/Total_Purchase,4)*100)+"%"
NonDisclosed_Percentage = str(round(NonDisclosed_Players/Total_Purchase,4)*100)+"%"

Gender_Demographics = pd.DataFrame({"Counts": [Male_Players,Female_Players,NonDisclosed_Players],
                                    "Percentage": [Male_Percentage,Female_Percentage,NonDisclosed_Percentage]},
                                    index=["Male","Female","Other / Non-Disclosed"])
Gender_Demographics
```


## Purchasing Analysis (Gender)

```python
#define bins range and group names
#in the original datafram, add new column "Age Category" to return age ranges for each player
#group the data by age catergory
#count by age group and calculte the percentage
#assign two colunmns to a new dataframe
#formate the new dataframe

bins = [0,9.5,14.5,19.5,24.5,29.5,34.5,39.5,200]
Catergory_Names = ["Under 10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "Over 40"]
purchase_data["Age Catergory"] = pd.cut(purchase_data["Age"], bins, labels=Catergory_Names)

Age_Demographics = purchase_data.groupby("Age Catergory")
ageCount = Age_Demographics["SN"].count()
agePercentage = (ageCount/Total_Purchase)*100
df_Age_Demographics = pd.DataFrame({"Total Count": ageCount,
                                  "Percentage of Players": agePercentage})

df_Age_Demographics["Percentage of Players"]=df_Age_Demographics["Percentage of Players"].map("{:.2f}%".format)
df_Age_Demographics
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group the data by players' names
#count the numbers of times they buy and calcute the total price for each player
#average price for each purchase 
#sort by total purchase value and formate the numbers

Names_Group = purchase_data.groupby(["SN"])

Purchase_by_Name = Names_Group["Purchase ID"].count()
Purchase_Value_Name = Names_Group["Price"].sum()
Average_Purchase_Value = (Purchase_by_Name / Purchase_by_Name)

Top_Spenders = pd.DataFrame({"Purchase Count": Purchase_by_Name,
                         "Average Purchase Price": Average_Purchase_Value,
                         "Total Purchase Value": Purchase_Value_Name})

Top_Spenders = Top_Spenders.sort_values("Total Purchase Value", ascending=False) 
Top_Spenders["Average Purchase Price"] = Top_Spenders["Average Purchase Price"].map("${:.2f}".format)
Top_Spenders["Total Purchase Value"] = Top_Spenders["Total Purchase Value"].map("${:.2f}".format)
Top_Spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group data by items ID and item names
#calculate and formate the numbers

Item_Group = purchase_data.groupby(["Item ID","Item Name"])

Purchase_Count_Item = Item_Group["SN"].count()
Total_Purchase_Value_Item = Item_Group["Price"].sum()
Item_Price = (Total_Purchase_Value_Item / Purchase_Count_Item)

Popular_Items1 = pd.DataFrame({"Purchase Count Item": Purchase_Count_Item,
                             "Item Price": Item_Price,
                              "Total Purchase Value": Total_Purchase_Value_Item})

Popular_Items1 = Popular_Items1.sort_values("Total Purchase Value", ascending=False) 
Popular_Items1
```
